```python
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)

var_name=os.path.basename(parentdir)

var_name
```

```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
```

```python
df_hh=pd.read_excel("../row_data/occ_rate.xlsx")
df_hh.head(5)
```

```python
unique_array = list(set(df_hh['Average household size (number of members)'].values))
str_in_array=[s for s in unique_array if type(s)==str]
str_in_array
```

```python
df_hh = df_hh.replace('..', np.nan)

# drop rows where all values are NaN
df_hh = df_hh.dropna(how='any')
df_hh
```

```python
unique_array = list(set(df_hh['Average household size (number of members)'].values))
str_in_array=[s for s in unique_array if type(s)==str]
str_in_array
```

```python
value = 2000
df_hh = df_hh[df_hh['Year'] >= value]
countries=df_countries['Country Name'].values
countries_hh=df_hh['Country'].values
if set(countries) == set(countries_hh):
    print("The arrays have the same elements.")
else:
    print("The arrays do not have the same elements.")
```

```python
mean=np.mean(df_hh['Average household size (number of members)'].values)
mean
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=range(2010,2021)
# mean=np.mean(df_hh['Average household size (number of members)'].values)


coun_year=[]
for count in zip(countries,ISO3):
    for year in years:
        coun_year.append([count[0],count[1],year])
df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
df_historical_out[var_name]=0

for count in df_historical_out['Nation'].values:
    
    if count in countries_hh:
        
        df_country = df_hh.loc[df_hh['Country']== count]
        
        vals = df_country['Average household size (number of members)'].values
        vlas = np.array(vals)
        dato = vlas.mean()
        
        for year in years:
            df_historical_out.loc[(df_historical_out['Nation'] == count) & 
                            (df_historical_out['Year']== year), var_name] = dato
            
    else:
        for year in years:
            df_historical_out.loc[(df_historical_out['Nation'] == count) & 
                        (df_historical_out['Year']== year), var_name] = mean
            
df_historical_out.head(9)
```

```python
df_historical_out['iso_code3'].unique(),len(df_historical_out['iso_code3'].unique())
```

```python
#### Projected ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years2=range(max(years)+1,2051)

coun_year=[]
for count in zip(countries,ISO3):
    for year in years2:
        coun_year.append([count[0],count[1],year])
df_projected_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
df_projected_out[var_name]=0
df_projected_out.head()
```

```python
for count,year in zip(df_projected_out['Nation'].values,df_projected_out['Year'].values):
    df_dato=df_historical_out[(df_historical_out['Nation']==count) & 
                               (df_historical_out['Year']==max(years))]
    dato=df_dato[var_name].values[0]
    df_projected_out.loc[(df_projected_out['Nation'] == count) & 
                        (df_projected_out['Year']== year), var_name] = dato
df_projected_out
```

```python
df_historical_out['Nation'].replace('Turkiye','Turkey',inplace=True)
df_projected_out['Nation'].replace('Turkiye','Turkey',inplace=True)

df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
