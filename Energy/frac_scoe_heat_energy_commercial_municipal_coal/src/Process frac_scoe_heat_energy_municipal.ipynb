```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
```

```python
df_WEBH2021=pd.read_excel('../row_data/WorldEnergyBalancesHighlights2021.xlsx',sheet_name=3)
```

```python
df_WEBH2021.columns = df_WEBH2021.iloc[0]
df_WEBH2021.replace(to_replace='..', value=0,inplace=True)
df_WEBH2021
```

```python
np.unique(df_WEBH2021['Flow'].values)
```

```python
df_sector1=df_WEBH2021[df_WEBH2021['Flow']=='Commercial and public services (PJ)']

df_sector2=df_WEBH2021[df_WEBH2021['Flow']=='Residential (PJ)']

df_total1=df_sector1[df_sector1['Product']=='Total']
df_total2=df_sector2[df_sector2['Product']=='Total']
```

```python
df_prop=pd.read_excel('../row_data/frac_scoe_heat_energy_proportion.xlsx')
df_prop
```

```python
df_prop_=df_prop[df_prop['SISEPUEDE_var']=='frac_scoe_heat_energy_commercial_municipal_coal']
df_prop_
```

```python
sisepuede_vars=[var for var in np.unique(df_prop['SISEPUEDE_var'].values)]
sisepuede_vars.append('frac_scoe_heat_energy_commercial_municipal_hydrogen')
sisepuede_vars
```

```python
#### HISTORICAL ######

nation_in_df=np.unique(df_WEBH2021['Country'].values)

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=range(2000,2020)

for s_var in sisepuede_vars:
    coun_year=[]
    df_historical_out=0
    for count in zip(countries,ISO3):
        for year in years:
            coun_year.append([count[0],count[1],year])
            
    df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
    
    df_historical_out[s_var]=0

    df_prop_var=df_prop[df_prop['SISEPUEDE_var']==s_var]
    
    for country in countries:
        for year in years:
        
        
            if country in nation_in_df:

                dato_fuel=[]

                for fuel,p in zip(df_prop_var['WEBH2021_var'].values,df_prop_var['Proportion'].values):
                    
                    #Sector 1

                    Total1=df_total1.loc[df_total1['Country']==country,[year]][year].values[0]

                    df_fuel1=df_sector1[df_sector1['Product']==fuel]

                    f1=df_fuel1.loc[df_fuel1['Country']==country,[year]][year].values[0]
                    
                    #Sector 2
                    
                    Total2=df_total2.loc[df_total2['Country']==country,[year]][year].values[0]

                    df_fuel2=df_sector2[df_sector2['Product']==fuel]

                    f2=df_fuel2.loc[df_fuel2['Country']==country,[year]][year].values[0]
                    
                    f = f1 + f2
                    
                    Total = Total1 + Total2 

                    dato_fuel.append(p*f/Total)

                    #print(dato_fuel)


                df_historical_out.loc[(df_historical_out['Nation'] == country) & 
                            (df_historical_out['Year']== year), s_var]=sum(dato_fuel)
            
            else:
                
                dato_fuel=[]
                
                for fuel,p in zip(df_prop_var['WEBH2021_var'].values,df_prop_var['Proportion'].values):
                    
                    #Sector 1

                    Total1=df_total1.loc[df_total1['Country']=='World',[year]][year].values[0]

                    df_fuel1=df_sector1[df_sector1['Product']==fuel]

                    f1=df_fuel1.loc[df_fuel1['Country']=='World',[year]][year].values[0]
                    
                    #Sector 2
                    
                    Total2=df_total2.loc[df_total2['Country']=='World',[year]][year].values[0]

                    df_fuel2=df_sector2[df_sector2['Product']==fuel]

                    f2=df_fuel2.loc[df_fuel2['Country']=='World',[year]][year].values[0]
                    
                    f = f1 + f2
                    
                    Total = Total1 + Total2 
                    
                    dato_fuel.append(p*f/Total)
                    
                df_historical_out.loc[(df_historical_out['Nation'] == country) & 
                                (df_historical_out['Year']== year), s_var] = sum(dato_fuel)
                
    
    df_historical_out.to_csv(grandparentdir+f'/{s_var}/input_to_sisepuede/historical/{s_var}.csv')
```

```python
#### Projected ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years2=range(max(years)+1,2051)

for s_var in sisepuede_vars:
    coun_year=[]
    for count in zip(countries,ISO3):
        for year in years2:
            coun_year.append([count[0],count[1],year])
            
    df_projected_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
    
    df_projected_out[s_var]=0

    #df_prop_var=df_prop[df_prop['SISEPUEDE_var']==s_var]
    
    df_historical=pd.read_csv(grandparentdir+f'/{s_var}/input_to_sisepuede/historical/{s_var}.csv')
    
    for count,year in zip(df_projected_out['Nation'].values,df_projected_out['Year'].values):
        
        df_dato=df_historical[(df_historical['Nation']==count) & 
                                   (df_historical['Year']==max(years))]
        dato=df_dato[s_var].values[0]
        
        df_projected_out.loc[(df_projected_out['Nation'] == count) & 
                            (df_projected_out['Year']== year), s_var] = dato
    
    
    df_projected_out.to_csv(grandparentdir+f'/{s_var}/input_to_sisepuede/projected/{s_var}.csv')
```
