```python
# Add GrandParent Folder to System Path

import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)

sys.path
```

```python
####

import pandas as pd
import numpy as np

#Variable name
var_name='occrate_trns_passenger_per_vehicle_public'
#DATAFRAMES
df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")

df_data_hist_rail=pd.read_csv (
grandparentdir+"/occrate_trns_passenger_per_vehicle_rail_passenger/input_to_sisepuede/historical/occrate_trns_passenger_per_vehicle_rail_passenger.csv")

df_data_proj_rail=pd.read_csv (
grandparentdir+"/occrate_trns_passenger_per_vehicle_rail_passenger/input_to_sisepuede/projected/occrate_trns_passenger_per_vehicle_rail_passenger.csv")

df_data_hist_bus=pd.read_csv (
grandparentdir+"/occrate_trns_passenger_per_vehicle_road_heavy_regional/input_to_sisepuede/historical/occrate_trns_passenger_per_vehicle_road_heavy_regional.csv")

df_data_proj_bus=pd.read_csv (
grandparentdir+"/occrate_trns_passenger_per_vehicle_road_heavy_regional/input_to_sisepuede/projected/occrate_trns_passenger_per_vehicle_road_heavy_regional.csv")
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=np.unique(df_data_hist_rail['Year'].values)
coun_year=[]
for count in zip(countries,ISO3):
    for year in years:
        coun_year.append([count[0],count[1],year])
df_historical_out = pd.DataFrame(coun_year, columns=['Country','iso_code3', 'Year'])
df_historical_out
```

```python
df_historical_out[var_name]=0
```

```python
for count,year in zip(df_historical_out['Country'].values,df_historical_out['Year'].values):
    df_rail=df_data_proj_rail[(df_data_proj_rail['Country'] == count) & (df_data_proj_rail['Year']== year)]
    data_value_rail=df_rail['occrate_trns_passenger_per_vehicle_rail_passenger'].values
    
    df_bus=df_data_proj_bus[(df_data_proj_bus['Country'] == count) & (df_data_proj_bus['Year']== year)]
    data_value_bus=df_bus['occrate_trns_passenger_per_vehicle_road_heavy_regional'].values
    
    
    df_historical_out.loc[(df_historical_out['Country'] == count) & 
                        (df_historical_out['Year']== year), var_name] = (data_value_rail+data_value_bus)/2
df_historical_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
#### Projected ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=np.unique(df_data_proj_rail['Year'].values)
coun_year=[]
for count in zip(countries,ISO3):
    for year in years:
        coun_year.append([count[0],count[1],year])
df_projected_out = pd.DataFrame(coun_year, columns=['Country','iso_code3', 'Year'])
df_projected_out
```

```python
df_projected_out[var_name]=0
```

```python
for count,year in zip(df_projected_out['Country'].values,df_projected_out['Year'].values):
    df_rail=df_data_proj_rail[(df_data_proj_rail['Country'] == count) & (df_data_proj_rail['Year']== year)]
    data_value_rail=df_rail['occrate_trns_passenger_per_vehicle_rail_passenger'].values
    
    df_bus=df_data_proj_bus[(df_data_proj_bus['Country'] == count) & (df_data_proj_bus['Year']== year)]
    data_value_bus=df_bus['occrate_trns_passenger_per_vehicle_road_heavy_regional'].values
    
    
    df_projected_out.loc[(df_projected_out['Country'] == count) & 
                        (df_projected_out['Year']== year), var_name] = (data_value_rail+data_value_bus)/2
df_projected_out
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
