```python
#Socioeconomic varibles

#GDP
df_gdp_hist = pd.read_csv(ggpdir+'/Socioeconomic/gdp_mmm_usd/input_to_sisepuede/historical/gdp_mmm_usd.csv')
df_gdp_proj = pd.read_csv(ggpdir+'/Socioeconomic/gdp_mmm_usd/input_to_sisepuede/projected/gdp_mmm_usd.csv')

nation_gdp = sorted(set(df_gdp_proj['Nation'].values))

years_all = range(min(df_gdp_hist['Year'].values),max(df_gdp_proj['Year'].values)+1)
c_y_gdp = []
for c in nation_gdp:
    for y in years_all:
    
        try:
            gdp = df_gdp_hist.loc[(df_gdp_hist['Nation']==c) & (df_gdp_hist['Year']==y),]['gdp_mmm_usd'].values[0]
            
        except:
            gdp = df_gdp_proj.loc[(df_gdp_proj['Nation']==c) & (df_gdp_proj['Year']==y)]['gdp_mmm_usd'].values[0]
            
        c_y_gdp.append([c,y,gdp])

df_gdp_all = pd.DataFrame(c_y_gdp,columns=['Nation', 'Year', 'gdp_mmm_usd'])



#Rural population
df_popru_hist = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_rural/input_to_sisepuede/historical/population_gnrl_rural.csv')
df_popru_proj = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_rural/input_to_sisepuede/projected/population_gnrl_rural.csv')

#Urban population
df_popur_hist = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_urban/input_to_sisepuede/historical/population_gnrl_urban.csv')
df_popur_proj = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_urban/input_to_sisepuede/projected/population_gnrl_urban.csv')

nation_pop = sorted(set(df_popur_proj['Nation'].values))

years_all = range(min(df_popur_hist['Year'].values),max(df_popur_proj['Year'].values)+1)
c_y_pop = []
for c in nation_pop:
    for y in years_all:
    
        try:
            pop_ur = df_popur_hist.loc[(df_popur_hist['Nation']==c) &
                                       (df_popur_hist['Year']==y),]['population_gnrl_urban'].values[0]
            pop_ru = df_popru_hist.loc[(df_popru_hist['Nation']==c) &
                                       (df_popru_hist['Year']==y),]['population_gnrl_rural'].values[0]
            
        except:
            pop_ur = df_popur_proj.loc[(df_popur_proj['Nation']==c) &
                                       (df_popur_proj['Year']==y),]['population_gnrl_urban'].values[0]
            pop_ru = df_popru_proj.loc[(df_popru_proj['Nation']==c) &
                                       (df_popru_proj['Year']==y),]['population_gnrl_rural'].values[0]
            
        c_y_pop.append([c,y,pop_ur,pop_ru])

df_pop_all = pd.DataFrame(c_y_pop,columns=['Nation', 'Year', 'population_gnrl_urban','population_gnrl_rural'])

#All Together
nation_pop_gdp = sorted(set(nation_pop).intersection(set(nation_gdp)))
years_all = range(min(df_popur_hist['Year'].values),max(df_popur_proj['Year'].values)+1)
c_y_pop_gdp = []
for c in nation_pop_gdp:
    iso3 = df_gdp_hist.loc[df_gdp_hist['Nation']==c]['iso_code3'].values[0]
    for y in years_all:
        pop_ur = df_pop_all.loc[(df_pop_all['Nation']==c) &
                                       (df_pop_all['Year']==y),]['population_gnrl_urban'].values[0]
        
        pop_ru = df_pop_all.loc[(df_pop_all['Nation']==c) &
                                       (df_pop_all['Year']==y),]['population_gnrl_rural'].values[0]
        
        gdp = df_gdp_all.loc[(df_gdp_all['Nation']==c) & (df_gdp_all['Year']==y)]['gdp_mmm_usd'].values[0]
        
        c_y_pop_gdp.append([c,iso3,y,pop_ur,pop_ru,gdp])
df_pop_gdp_all = pd.DataFrame(c_y_pop_gdp,columns=['Nation','iso_code3', 'Year',
                                                   'population_gnrl_urban',
                                                   'population_gnrl_rural',
                                                   'gdp_mmm_usd'])
df_pop_gdp_all
```

```python
nation_enduse = df_energy_building['Country'].values
nation_pop_gdp_all=df_pop_gdp_all['Nation'].unique()

nation_pop_gdp_enduse = sorted(set(nation_pop_gdp_all).intersection(set(nation_enduse)))
years_hist = range(2011,2019+1)

c_y = []
for c in nation_pop_gdp_enduse:
    for y in years_hist:
        c_y.append([c,y])

df_energy_products=pd.DataFrame(c_y,columns=['Nation','Year'])
      
for product in Products:
    
    df_energy_products[product]=0
    
    for c in nation_pop_gdp_enduse:
        for y in years_hist:
            prod_val=0
            try:
                prod_val+=df_energy_building[(df_energy_building['Product'].str.startswith(product))&
                          (df_energy_building['Country']==c)][y].values[0]
            # some End Use does not have all products specified, those are taken to be 0 by passing
            except:
                pass
            df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),product]=prod_val
    
df_energy_products
```

```python
for c in df_energy_products['Nation'].unique():
    for y in df_energy_products['Year'].unique():
        for product in Products:
            prod_val=df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),'Total final energy use (PJ)'].values[0]
            if prod_val==0:
                val=np.mean(df_energy_products[df_energy_products['Nation']==c][product].values)
                
                df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),product]=val
df_energy_products
```

```python
# Cross Walk with IEA Fuel products

# frac_scoe_X_coal : [Coal and coal products (PJ)      ]
#  frac_scoe_X_diesel : [0.6*Oil and oil products (PJ)      ]
#  frac_scoe_X_electricity : [Electricity (PJ)      ]
#  frac_scoe_X_gasoline :  [0.2*Oil and oil products (PJ)      ]
#  frac_scoe_X_hydrocarbon_gas_liquids :  [0.2*Oil and oil products (PJ)      ]
#  frac_scoe_X_hydrogen : 0
#  frac_scoe_X_kerosene:  [Other sources (PJ)        ]
#  frac_scoe_X_natural_gas : [Gas (PJ) + Heat (PJ)              ]
#  frac_scoe_X_solid_biomass : [Biofuels and waste (PJ)   ]

# 'frac_scoe_heat_energy_other_se_coal',
# 'frac_scoe_heat_energy_other_se_diesel',
# 'frac_scoe_heat_energy_other_se_electricity',
# 'frac_scoe_heat_energy_other_se_gasoline',
# 'frac_scoe_heat_energy_other_se_hydrocarbon_gas_liquids',
# 'frac_scoe_heat_energy_other_se_hydrogen',
# 'frac_scoe_heat_energy_other_se_kerosene',
# 'frac_scoe_heat_energy_other_se_natural_gas',
# 'frac_scoe_heat_energy_other_se_solid_biomass'

cross_walk_dict={'frac_scoe_heat_energy_other_se_coal':[(1,'Coal and coal products (PJ)')],
'frac_scoe_heat_energy_other_se_diesel': [(0.6,'Oil and oil products (PJ)')],
'frac_scoe_heat_energy_other_se_electricity':[(1,'Electricity (PJ)')],
'frac_scoe_heat_energy_other_se_gasoline':[(0.2,'Oil and oil products (PJ)')],
'frac_scoe_heat_energy_other_se_hydrocarbon_gas_liquids':[(0.2,'Oil and oil products (PJ)')],
'frac_scoe_heat_energy_other_se_hydrogen':[(0,'Oil and oil products (PJ)')],
'frac_scoe_heat_energy_other_se_kerosene':[(0.4,'Other sources (PJ)')],
'frac_scoe_heat_energy_other_se_natural_gas': [(1,'Gas (PJ)'),(1,'Heat (PJ)')],
'frac_scoe_heat_energy_other_se_solid_biomass':[(1,'Biofuels and waste (PJ)')]}

for c in df_energy_products['Nation'].unique():
    for y in df_energy_products['Year'].unique():
        
        total_energy=0
        
        for ssp_var in cross_walk_dict.keys():    
            for (frac,product) in cross_walk_dict[ssp_var]:
                
                total_energy+=frac*df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),product].values[0]
                
        df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),'Pseudo Total (PJ)']=total_energy
        
        for ssp_var in cross_walk_dict.keys():
            var_energy=0
            for (frac,product) in cross_walk_dict[ssp_var]:
                
                var_energy+=frac*df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),product].values[0]
            total_ener=df_energy_products.loc[(df_energy_products['Nation']==c)&
                                  (df_energy_products['Year']==y),'Pseudo Total (PJ)'].values[0]
            if  total_ener!=0:   
                df_energy_products.loc[(df_energy_products['Nation']==c)&
                              (df_energy_products['Year']==y),ssp_var]=var_energy/total_ener
            else:
                df_energy_products.loc[(df_energy_products['Nation']==c)&
                              (df_energy_products['Year']==y),ssp_var]=0
df_energy_products
```
